```python
## start logistic regression exercise
data = pd.read_csv('BC Data new.csv')
data.head()
```

```python
data_train=data[0:2000]
import statsmodels.api as sm
import statsmodels.formula.api as smf

formula = 'choice ~ male + buyamt + freq + lastbuy + firstbuy'
result = smf.logit(formula = str(formula), data = data_train).fit()
result.summary()
```

```python
data['prob'] = pd.DataFrame(result.predict(data))
data.head()
```

```python
data['pred'] = [0 if x<  .5  else 1 for x in data['prob']]
data.head()
```

```python
from sklearn.metrics import accuracy_score
data_in=data[0:2000]
accuracy_score(data_in['choice'], data_in['pred'])
```

```python
data_out = data[2000:]
accuracy_score(data_out['choice'], data_out['pred'])
```

```python
accuracy_score(data['choice'], data['pred'])
```
